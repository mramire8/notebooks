## Methods per Trial

The data comes from the methods logs, includes accuracy and auc. This notebook only uses accuracy at the momment. The methods included are: 

* RND: random sentence
* First1: first sentence
* SR: SR with calibration
* True: where the oracle is the true oracle of the data

All methods are based on RND+Strategy (sequential). 



```python
from collections import defaultdict
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.style.use('bmh')


def to_float(x):
    f = x
    try:
        f = float(x)
    except:
        f = x
    return f

def read_data(filename, measure='accu'):
    if measure == 'accu':
        col = 1  ## col=0 is the cost, 1 accuracy and 2 auc, rest of columns are confusion matrix of the oracle
    elif measure =='auc':
        col = 2
    else:
        raise Exception("Oops, we cant do that measure yet.")
        
    f = open(filename)
    with f:
        lines = f.readlines() 
    data =  [l.strip().split("\t") for l in lines[2:]] #discard first 2 lines
    data = np.array([[to_float(x) for x in row] for row in data])
    data_xy =  defaultdict(lambda: [])
    for d in data:
        data_xy[d[0]].append(d[col])
    return data_xy

def get_trial(data, t):
    cost = range(10, 3000, 10)
    trial = []
    for c in cost:
        trial.append(data[c][t])
    return trial

path = './logs'
methods = {'rnd':'imdb_rnd_rnd', 'sr':'imdb_rnd_sr_cal', 'first1':'imdb_rnd_first1', 'true':'imdb_rnd_true'}

data = {}
for m,f in methods.items():
    data[m] = read_data(path +'/' + methods[m] + '.log')
    
```

```python
# print len(data['rnd'])
# print data['rnd'][20]
# x=  get_trial(data['rnd'], 4)
# print x[:5]

def plot_trial_methods(methods, data, t):
    y_data = []
    for m in methods:
        y_data.append(get_trial(data[m], t))
    cost = range(10, 3000, 10)
    for y, name in zip(y_data, methods):
        plt.plot(cost, y, label=name)
    plt.legend(loc='best')
    plt.title('Accuracy on Trial ' + str(t))
    plt.ylabel('Accuracy')
    plt.xlabel('Cost')
    plt.show()
    
```

### Plot per Trial of all Methods

####Notes:

* In some trials SR is comparable with other methods
* Most trials (4/5) SR is the worst


```python
for t in range(5):
    plot_trial_methods(methods.keys(), data, t)
```

### Method plot with all trials

The following plots show one method per plot where each curve is a trial in the data. RND and Sr are of interestin these plots. 

####Notes:

* Compared to RND, SR has a larger variance per trial. However, two out of five trials are comparable to RND. 
* Among the trial RND doesn't have big changes in performance.

```python
def plot_method(method, data):
    for t in range(5):
        cost = range(10, 3000, 10)
        y = get_trial(data[method],t)
        plt.plot(cost, y, label='trial '+str(t))
    plt.legend(loc='best')
    plt.title('Accuracy of '+method.upper())
    plt.xlabel('Cost')
    plt.ylabel('Accuracy')
    plt.show()

plot_method('rnd', data)
```
