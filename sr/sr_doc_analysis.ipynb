```python
import utilities.experimentutils as exputil
import learner
import utilities.datautils as datautil
import numpy as np
import experiment.base as exp
import nltk
```

```python
## Get the data ready
imdb_path = 'C:/Users/mramire8/Documents/Research/Oracle confidence and Interruption/dataset/aclImdb/raw-data'
imdb_path = '/Users/maru/MyCode/data/imdb'

rnd = np.random.RandomState(2345)
clf = exputil.get_classifier('lrl2',parameter=1)
expert = exputil.get_classifier('lrl2',parameter=1)
vct = exputil.get_vectorizer({'vectorizer':"tfidf", 'limit':None, 'min_size':None})
data = datautil.load_dataset('imdb', imdb_path, categories=None, rnd=5463, shuffle=True)
data.train.data = np.array(data.train.data, dtype=object)
data.test.data = np.array(data.test.data, dtype=object)
data.train.bow = vct.fit_transform(data.train.data)
data.test.bow = vct.transform(data.test.data)
expert = exputil.get_classifier('lrl2',parameter=1)
```

```python
## Set the learner options and expert
sent_tk = nltk.data.load('tokenizers/punkt/english.pickle')
st = learner.strategy.StructuredLearner(clf)
st.set_sent_tokenizer(sent_tk)
st.set_vct(vct)
st.set_snippet_utility('sr')
st.set_calibration(True)
expert.fit(data.train.bow, data.train.target)
```

```python
## Get the boostrap and train
data.train.remaining = rnd.permutation(len(data.train.target))
## balanced bootstrap
# bt_obj = learner.strategy.BootstrapFromEach(None, seed=5463)
# initial = bt_obj.bootstrap(data.train, step=200, shuffle=True)
## random bootstrap 
initial = rnd.choice(data.train.remaining, 200, replace=False)
st.fit(data.train.bow[initial], data.train.target[initial], doc_text=data.train.data[initial])
```

```python
## Select N random documents
n = 100
rnd_docs = list(set(data.train.remaining) - set(initial))
np.random.shuffle(rnd_docs)
print len(rnd_docs), rnd_docs[:10]
```

```python
# Process
## Get the sentences per document
sent = sent_tk.tokenize_sents(data.train.data[rnd_docs[:n]])
## compute the max scoring snippet (with calibration)
sent_scores, snippet_text = st._compute_snippet(data.train.data[rnd_docs[:n]])

# #print scores per sentence
## DOCID, TARGET(True label), SENTENCE ID, PRED (student), SCORE(maxprob), Text
d = 0
for i, idx in enumerate(rnd_docs[:100]):
    
    print "-"*40
    print "Doc:", i
    print "Target:", data.train.target[idx] 
    bow = vct.transform(sent[i])
    pred = st.snippet_model.predict(bow)
    exp_pred = expert.predict(bow)
    print "ID\tPred\tExpertPred\tScore\ttext"
    for j, s in enumerate(zip(sent[i], pred, exp_pred)):
        if snippet_text[i] == s[0]:  # if selected sentence, mark it
            print "*%s\t%s\t%s\t%.3f\t%s" % (j, s[1], s[2], st._snippet_max(bow[j]), s[0])
        else:
            print "%s\t%s\t%s\t%.3f\t%s" % (j, s[1], s[2], st._snippet_max(bow[j]), s[0])
        d +=1
    print sent_scores[i], snippet_text[i]
    if i > 100:
        break
print d
```

```python

def print_document_senteces(data, st, indices, n=100):
    ## compute the max scoring snippet (with calibration)
    sent_scores, snippet_text = st._compute_snippet(data.data[indices[:n]])

    sent = sent_tk.tokenize_sents(data.data[indices[:n]])
    results = []
    for i, idx in enumerate(indices[:n]):
        doc_i = []
        bow = vct.transform(sent[i])
        pred = st.snippet_model.predict(bow)
        exp_pred = expert.predict(bow)
        for j, s in enumerate(zip(sent[i], pred, exp_pred)):
            selected = False
            if snippet_text[i] == s[0]:
                selected = True
            doc_i.append([i, data.target[idx], j, s[1], s[2],st._snippet_max(bow[j]), s[0], selected])
        results.extend(doc_i)
    return results


# Doc, Target, ID,	Pred,	ExpertPred,	Score,	text

r =  print_document_senteces(data.train, st, rnd_docs)
# no_m = [x for x in r if ]
print "Total:", len(r)
print "Student Pred 0:",len([x for x in r if x[3] == 0])
print "True 0:", len([x for x in r if x[1] == 0])
print "Expert Pred 0:",len([x for x in r if x[4] == 0])
print "Boostrap: ", data.train.target[initial].sum()
```

## Test Multiple Bootstrap 

* Get different bootstrap of size 200
* Get a 100 document test (fixed for all student test)
* For each student test the accuracy on all sentences in the test
* For each student check the accuracy on the selected sentence
* For each test check the accuracy of the oracle on all sentences (this should no change over the different student tests)
* For each test chcek the accuracy of the oracle on selected sentences

```python
def print_accu_all_sentences(r):
    print "Expert Accu on rnd docs: %.3f (%s / %s)" % (accuracy_score([x[1] for x in r], [x[4] for x in r]), len([x for x in r if x[4] == x[1]]), len(r))
    print "Student Accu on rnd docs:  %.3f (%s / %s)" % (accuracy_score([x[1] for x in r], [x[3] for x in r]) ,len([x for x in r if x[3] == x[1]]), len(r))

    print "CM Student"
    cms = confusion_matrix([x[1] for x in r], [x[3] for x in r])
    print print_cm(cms)
    
    # print confusion_matrix(data.test.target, test)
    print "CM Expert"
    cme = confusion_matrix([x[1] for x in r], [x[4] for x in r])
    print print_cm(cme)
    
    return cms, cme

def print_accu_selected_sentences(r):
    f = [s for s in r if s[7] == True]
    cms, cme = print_accu_all_sentences(f)
    return cms, cme

def print_cm(cm):
    return "\n".join(["{}\t{}".format(*r) for r in cm])

#Get random document for testing (will not change over time)
test_idx = rnd.permutation(len(data.test.target))

n = len(data.train.target)
all_cms = np.zeros((2,2))
all_cme = np.zeros((2,2))
sel_cms = np.zeros((2,2))
sel_cme = np.zeros((2,2))

for i in range(5):
    train_idx = rnd.choice(range(n), 200, replace=False)
    print "-" * 40
    print "\nTrial ", i
    # Train a student on rnadom docs
    st.fit(data.train.bow[train_idx], data.train.target[train_idx], doc_text=data.train.data[train_idx])
    # Get sentences and predictions
    r =  print_document_senteces(data.test, st, test_idx, n=100)
    
    #Print results
    print "ALL:"
    cms, cme = print_accu_all_sentences(r)
    all_cms += cms
    all_cme += cme # this matrix should be the same for all trials
    print "SELECTED:"
    cms, cme = print_accu_selected_sentences(r)
    sel_cms += cms
    sel_cme += cme
```

### Analysis of CM

What is the average of class distribution?

* On all sentences the expert is balanced in average (see results below) 
* On all sentences the student tends to be imbalanced towards class 0, even though some trials are more class 1
* On selected sentences the student is imabalaced towards class 0 (same as all sentences)
* On selected sentences the expert is not as balaced 

```python
# Average Confusion Matrix of Student and Expert
print "On all Sentence"
print "Student:\n", print_cm(all_cms / 5.)
print "Avg. class dist: %.3f, %.3f" % (((all_cms[0][0] + all_cms[1][0])/5.)/(all_cms.sum()/5.), 
                                       ((all_cms[0][1] + all_cms[1][1])/5.) / (all_cms.sum()/5.))
print "Expert:\n", print_cm(all_cme / 5.)
print "Avg. class dist: %.3f, %.3f" % (((all_cme[0][0] + all_cme[1][0])/5.)/(all_cme.sum()/5.), 
                                       ((all_cme[0][1] + all_cme[1][1])/5.) / (all_cme.sum()/5.))

print"\nOn selected only"
print "Student:\n", print_cm(sel_cms / 5.)
print "Avg. class dist: %.3f, %.3f" % (((sel_cms[0][0] + sel_cms[1][0])/5.) / (sel_cms.sum()/5.), 
                                       ((sel_cms[0][1] + sel_cms[1][1])/5.) / (sel_cms.sum()/5.))

print "Expert:\n", print_cm(sel_cme / 5.)
print "Avg. class dist: %.3f, %.3f" % (((sel_cme[0][0] + sel_cme[1][0])/5.) / (sel_cme.sum()/5.), 
                                       ((sel_cme[0][1] + sel_cme[1][1])/5.) / (sel_cme.sum()/5.))
```

```python
def run_cm_test(data, bt=200):
    #Get 100 document for testing (will not change over time)
    test_idx = rnd.permutation(len(data.test.target))

    n = len(data.train.target)
    all_cms = np.zeros((2,2))
    all_cme = np.zeros((2,2))
    sel_cms = np.zeros((2,2))
    sel_cme = np.zeros((2,2))

    for i in range(5):
        train_idx = rnd.choice(range(n), bt, replace=False)
        print "-" * 40
        print "\nTrial ", i
        # Train a student on rnadom docs
        st.fit(data.train.bow[train_idx], data.train.target[train_idx], doc_text=data.train.data[train_idx])
        # Get sentences and predictions
        r =  print_document_senteces(data.test, st, test_idx, n=100)

        #Print results
        print "ALL:"
        cms, cme = print_accu_all_sentences(r)
        all_cms += cms
        all_cme += cme # this matrix should be the same for all trials
        print "SELECTED:"
        cms, cme = print_accu_selected_sentences(r)
        sel_cms += cms
        sel_cme += cme
        
    print "\nOn all Sentence"
    print "Student:\n", print_cm(all_cms / 5.)
    print "Avg. class dist: %.3f, %.3f" % (((all_cms[0][0] + all_cms[1][0])/5.)/(all_cms.sum()/5.), 
                                           ((all_cms[0][1] + all_cms[1][1])/5.) / (all_cms.sum()/5.))
    print "Expert:\n", print_cm(all_cme / 5.)
    print "Avg. class dist: %.3f, %.3f" % (((all_cme[0][0] + all_cme[1][0])/5.)/(all_cme.sum()/5.), 
                                           ((all_cme[0][1] + all_cme[1][1])/5.) / (all_cme.sum()/5.))

    print"\nOn selected only"
    print "Student:\n", print_cm(sel_cms / 5.)
    print "Avg. class dist: %.3f, %.3f" % (((sel_cms[0][0] + sel_cms[1][0])/5.) / (sel_cms.sum()/5.), 
                                           ((sel_cms[0][1] + sel_cms[1][1])/5.) / (sel_cms.sum()/5.))

    print "Expert:\n", print_cm(sel_cme / 5.)
    print "Avg. Accu.", (sel_cme[0][0] + sel_cme[1][1]) / sel_cme.sum() / 5.
    print "Avg. class dist: %.3f, %.3f" % (((sel_cme[0][0] + sel_cme[1][0])/5.) / (sel_cme.sum()/5.), 
                                           ((sel_cme[0][1] + sel_cme[1][1])/5.) / (sel_cme.sum()/5.))



## Try 200 bootstrap, sencond time
run_cm_test(data, bt=200)
```
